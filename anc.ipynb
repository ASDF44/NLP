```python
# import webtext from nltk.corpus
from nltk.corpus import webtext
```

```python
# load all fileids name into list x
x = webtext.fileids()
x
```

```python
# iterate over fileids to get first 20 characters of each fileid
for book in x:
    print(webtext.words(fileids=book)[:20])
    
```
