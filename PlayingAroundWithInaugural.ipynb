```python
# n-gram generation

from nltk.util import ngrams
# function to get n-grams from sentences.
def extract_ngrams(data, num):
    nGrams = ngrams(data, num)
    return [ ' '.join(grams) for grams in nGrams]
```

```python
# finding the most common n-grams among the presidential speeches
for i in range(2,4):
    d = []
    for term in inaugural.fileids():
        speech = inaugural.words(fileids = [term])
        temp = extract_ngrams(speech, i)
        for j in temp:
            d.append(j)
d
```

```python
repo = {}
for i in d:
    if i not in repo.keys():
        repo[i] = 0
    repo[i] += 1
repo
```
